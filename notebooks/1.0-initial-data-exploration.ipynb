# Data Exploration

Firstly, let's look at the whole dataset (u.data)

```python
# All required imports are in this cell

import pandas as pd
```

```python
# Dataset root folder
dataset_root_path = "../data/interim/ml-100k"
```

```python
# u.data doesn't contain a header, so we add it ourselves using the dataset description 
u_data_df = pd.read_csv(f"{dataset_root_path}/u.data", delimiter="\t", names=["user_id", "item_id", "rating", "timestamp"])
```

```python
u_data_df.head()
```

```python
u_info_df = pd.read_csv(f"{dataset_root_path}/u.info", delimiter="\t", names=["number_of"])
u_info_df.head()
```

```python
# pandas cannot read this csv if the encoding is not specified
# `file -I u.item` gives us us-ascii, however, it also does not work
# cp1252 works

u_item_df = pd.read_csv(f"{dataset_root_path}/u.item", delimiter="|", names=["movie_id", "movie_title", "release_date",
                                                                              "video_release_date", "IMDb_URL",
                                                                              "unknown", "Action", "Adventure", "Animation", "Childrens",
                                                                              "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir",
                                                                              "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller",
                                                                              "War", "Western"], encoding="cp1252")
u_item_df.head()
```
